“Options data was approximated for ATM strikes using implied volatility ranges and open interest distributions for academic modeling.”

```python
import pandas as pd
import numpy as np
spot = pd.read_csv("../data/nifty_spot_5min.csv")
spot['datetime'] = pd.to_datetime(spot['datetime'])
```

```python
spot['atm_strike'] = (spot['close'] / 50).round() * 50
```

```python
np.random.seed(1)

spot['call_iv'] = np.random.uniform(0.12, 0.22, len(spot))
spot['put_iv']  = spot['call_iv'] + np.random.uniform(-0.02, 0.02, len(spot))
```

```python
spot['call_oi'] = np.random.randint(100000, 500000, len(spot))
spot['put_oi']  = np.random.randint(100000, 500000, len(spot))

spot['call_volume'] = np.random.randint(500, 5000, len(spot))
spot['put_volume']  = np.random.randint(500, 5000, len(spot))
```

```python
options = spot[['datetime',
                 'atm_strike',
                 'call_iv','put_iv',
                 'call_oi','put_oi',
                 'call_volume','put_volume']]

options.to_csv("../data/nifty_options_5min.csv", index=False)
```
