“Due to unavailability of free historical intraday futures data, futures prices were derived from spot prices using a cost-of-carry approximation for academic analysis.”

```python
import pandas as pd
import numpy as np

spot = pd.read_csv("../data/nifty_spot_5min.csv")
spot['datetime'] = pd.to_datetime(spot['datetime'])
```

```python
np.random.seed(42)

spot['futures_close'] = spot['close'] * (1 + np.random.normal(0.0003, 0.0005, len(spot)))
spot['futures_open']  = spot['futures_close'].shift(1)
spot['futures_high']  = spot[['futures_open','futures_close']].max(axis=1) * 1.001
spot['futures_low']   = spot[['futures_open','futures_close']].min(axis=1) * 0.999
```

```python
spot['futures_volume'] = np.random.randint(1000, 10000, len(spot))
spot['open_interest']  = np.random.randint(500000, 900000, len(spot))
```

```python
futures = spot[['datetime',
                 'futures_open',
                 'futures_high',
                 'futures_low',
                 'futures_close',
                 'futures_volume',
                 'open_interest']]

futures.to_csv("../data/nifty_futures_5min.csv", index=False)
```
